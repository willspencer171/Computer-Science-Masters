# FP-Growth Algorithm in Python

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
data = [['Milk', 'Burgers', 'Buns', 'Ketchup'],
        ['Burgers', 'Buns', 'Ketchup', 'Coals'],
        ['Milk', 'Buns', 'Beer'],
        ['Burgers', 'Beer', 'Buns', 'Coals'],
        ['Milk', 'Ketchup'],
        ['Milk', 'Beer', 'Burgers', 'Buns', 'Ketchup', 'Coals'],
        ['Beer', 'Burgers', 'Buns', 'Ketchup', 'Coals'],
        ['Milk', 'Buns'],
        ['Burgers', 'Buns', 'Ketchup'],
        ['Beer', 'Burgers', 'Ketchup', 'Coals']]
```

```python
# Encode data as market basket data
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)

df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
fp = fpgrowth(df, min_support=0.25, use_colnames=True)
```

```python
assoc = association_rules(fp, min_threshold=.5)
assoc[assoc['consequents'].apply(lambda x: 'Coals' in x)]
```
